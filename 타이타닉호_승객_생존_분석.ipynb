## 타이타닉호 승객 생존 분석

### 개요

본 분석에서는 seaborn 라이브러리에서 제공하는 'titanic' 데이터셋을 사용하여 타이타닉호 승객의 생존에 영향을 미치는 요인을 분석합니다. 객실 등급, 성별, 나이 등 다양한 변수를 고려하여 생존율을 비교하고, 변수 간의 상관관계를 분석합니다.

### 데이터셋

- `seaborn.load_dataset('titanic')` 함수를 사용하여 'titanic' 데이터셋을 불러옵니다.
- 데이터셋은 타이타닉호 승객의 정보를 담고 있으며, 생존 여부, 객실 등급, 성별, 나이, 승선 위치 등의 변수를 포함합니다.

### 분석 내용

1. **객실 등급별 생존율 비교**: `sns.barplot`을 사용하여 객실 등급에 따른 생존율을 시각화합니다.
2. **성별에 따른 생존율 비교**: `sns.barplot`을 사용하여 성별에 따른 생존율을 시각화합니다.
3. **나이와 생존 여부의 관계**: `sns.histplot`을 사용하여 나이와 생존 여부의 관계를 시각화합니다.
4. **객실 등급, 성별, 나이를 고려한 생존율 비교**: `sns.catplot`을 사용하여 객실 등급, 성별, 나이를 함께 고려한 생존율을 시각화합니다.
5. **상관관계 분석**: `sns.heatmap`을 사용하여 변수 간의 상관관계를 히트맵으로 시각화합니다.

### 결론

본 분석을 통해 타이타닉호 승객의 생존에 영향을 미치는 다양한 요인을 확인할 수 있습니다. 객실 등급, 성별, 나이 등의 변수가 생존율과 관련이 있으며, 이러한 분석 결과는 사고 당시의 상황과 사회적 배경을 이해하는 데 도움을 줄 수 있습니다.

### 코드

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 데이터셋 불러오기
titanic = sns.load_dataset('titanic')

# 데이터 확인
print(titanic.head())
print(titanic.info())

# 데이터 분석 및 시각화

# 객실 등급별 생존율 비교
sns.barplot(x='class', y='survived', data=titanic)
plt.title('객실 등급별 생존율')
plt.show()

# 성별에 따른 생존율 비교
sns.barplot(x='sex', y='survived', data=titanic)
plt.title('성별에 따른 생존율')
plt.show()

# 나이와 생존 여부의 관계
sns.histplot(x='age', hue='survived', data=titanic)
plt.title('나이와 생존 여부의 관계')
plt.show()

# 객실 등급, 성별, 나이를 고려한 생존율 비교
sns.catplot(x='class', y='survived', hue='sex', col='alive', kind='bar', data=titanic)
plt.show()

# 상관관계 분석 (heatmap)
corr_matrix = titanic.corr()
sns.heatmap(corr_matrix, annot=True)
plt.title('상관관계 분석')
plt.show()
```
